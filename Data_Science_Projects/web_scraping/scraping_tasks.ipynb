# Task 1

```python
import requests

res = requests.get('https://www.xe.com/currencyconverter/convert/?Amount=1&From=USD&To=EGP')
res
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(res.text, 'html.parser')
egp_text = soup.find('p', {'class': 'sc-b5d1d4ba-1 bPeOTN'}).text
egp_text
```

```python
egp = float(egp_text.split(' ')[0])
egp
```

##### program to convert dollar to egyptian pound

```python
dollars = float(input('Enter the amount of dollars: '))
egp_amount = dollars * egp
print(f'{dollars} dollars is equal to {egp_amount} EGP')
```

# Task 2

```python
res = requests.get('https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168')
res
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(res.text, 'html.parser')
forecast = soup.find_all('div', {'class': 'tombstone-container'})
forecast
```

```python
with open('forecast.html', mode = 'w', encoding= 'utf-8') as f:
    f.write('Day, Description, status, Temperature\n')
    for item in forecast:
        day = item.find('p', {'class': 'period-name'}).text
        desc = item.find('p', {'class': 'short-desc'}).text
        temp_content = item.find('p', {'class': 'temp'}).text
        status = temp_content.split(':')[0]
        temp = temp_content.split(':')[1]
        f.write(f'{day}, {desc}, {status}, {temp}\n')
```

```python
import pandas as pd
df = pd.read_csv('forecast.html')
df
```

# Task 3

```python
res = requests.get('https://books.toscrape.com/')
res
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(res.text, 'html.parser')


```

```python
pages_number = int(soup.find('li', {'class': 'current'}).text.strip().split(' ')[3])
pages_number
```

```python
with open ('books.csv', mode = 'w', encoding = 'utf-8') as f:
    f.write('Title, Price, rate\n')
    for i in range(pages_number):
        res = requests.get(f'https://books.toscrape.com/catalogue/page-{i+1}.html')
        soup = BeautifulSoup(res.text, 'html.parser')
        books = soup.find_all('article', {'class': 'product_pod'})
        for book in books:
            title = book.find('h3').find('a').text
            price = book.find('p', {'class': 'price_color'}).text
            rate = book.find('p', {'class': 'star-rating'}).get('class')[1]
            f.write(f'"{title}", "{price}", "{rate}"\n')
        print(f'Page {i+1} done')
```

```python
import pandas as pd
df = pd.read_csv('books.csv')
df
```

# Task 4

```python
res = requests.get('https://www.scrapethissite.com/pages/simple/')
res
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
with open('countries.csv', mode = 'w', encoding = 'utf-8') as fd:
    fd.write('name, capital, population, area\n')
    for item in soup.find_all('div', {'class': 'col-md-4 country'}):
        name = item.find('h3',{'class':'country-name'}).text.replace('\n','').replace('\r','')
        capital = item.find('span',{'class','country-capital'}).text
        population = item.find('span',{'class','country-population'}).text
        area = item.find('span',{'class','country-area'}).text
        fd.write(f'"{name}", "{capital}", "{population}", "{area}"\n')
   
```

```python
pd.read_csv('countries.csv')
```

# Task 5

```python
res = requests.get('https://www.scrapethissite.com/pages/forms/?page_num=1')
res
```

```python
soup = BeautifulSoup(res.text, 'html.parser')

with open('teams.csv', mode='w', encoding='utf-8') as fd:
    fd.write('Team Name, Year, Wins, Losses, Win %, Goals For (GF), Goals Against (GA)\n')
    for item in soup.find_all('tr', {'class': 'team'}):
        team_name = item.find('td', {'class': 'name'}).text.strip()
        year = item.find('td', {'class': 'year'}).text.strip()
        wins = item.find('td', {'class': 'wins'}).text.strip()
        losses = item.find('td', {'class': 'losses'}).text.strip()
        win_percentage = item.find('td', {'class': 'pct'}).text.strip()
        goals_for = item.find('td', {'class': 'gf'}).text.strip()
        goals_against = item.find('td', {'class': 'ga'}).text.strip()
        fd.write(f'"{team_name}", "{year}", "{wins}", "{losses}", "{win_percentage}", "{goals_for}", "{goals_against}"\n')  
```

```python
pd.read_csv('teams.csv')
```

```python
res = requests.get('https://jsonplaceholder.typicode.com/users')
res
```

```python
content = res.json()
content
```

```python
with open ('users.csv', mode = 'w', encoding='utf-8') as fd:
    fd.write('name, username, email, street, suite, city, zip code, geo_lat, geo_long\n')
    for item in content:
        name = item['name']
        username = item['username']
        email = item['email']
        street = item['address']['street']
        suite = item['address']['suite']
        city = item['address']['city']
        zipCode = item['address']['zipcode']
        lat = item['address']['geo']['lat']
        lng = item['address']['geo']['lng']
        fd.write(f'{name}, {username},{email},{street},{suite},{city},{zipCode},{lat},{lng}\n')
```

```python
pd.read_csv('users.csv')
```
